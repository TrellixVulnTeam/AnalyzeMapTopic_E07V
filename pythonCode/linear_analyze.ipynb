```python
import pprint
x=[]
for i in set(wowDfData.Called):
    y={}
    N=len(wowDfData[(wowDfData.Called==i)])
    corData=tuple(wowDfData[(wowDfData.Called==i)].corr()['avgDailyNet'][['costPower_Analyze','Nhuman_Analyze']])
    y['name']=i
    y['ads_costPower']=round(corData[0],3)
    y['ads_Nhuman']=round(corData[1],3)
    y['N']=N
    x.append(y)
pprint.pprint(x)
```

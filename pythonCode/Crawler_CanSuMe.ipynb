```python
import requests
from bs4 import BeautifulSoup as bs
import time
import geocoder

pages=68

diens=[]
for page in range(68):
    res=requests.get("https://www.cosmed.com.tw/Contact/Shop_list.aspx?pg={}&City=&District=&ID=&KeyWord=&Shop=".format(page+1))
    if res.status_code!=200:
        continue
    soup=bs(res.text,'lxml')
    for i in soup.select("div.detailNo > div.outletsList"):
        dien={}
        for j in i.select("table > tr"):
            dien[j.select("td")[0].text]=j.select("td")[1].text.replace("\n","")
        diens.append(dien)
    print('已完成{}頁'.format(page+1))
    time.sleep(1)
    

for i in range(2):
    for i in diens:
        try:
            if 'lat' not in i:
                i['lat'],i['lng']=geocoder.arcgis(i['門市地址'].split("號")[0].split("-")[0].split("、")[0].split("~")[0]).latlng
                time.sleep(4)
                print(i['lat'],i['lng'])
        except:
            pass
    
import re
for dien in diens:
    dien['address']=dien.pop('門市地址')
    dien['name']=dien.pop('門市名稱')
    try:
        dien["bigadd"] = re.findall("(\w\w[市|縣])(\w\w?\w?[區|市|鎮|鄉])",dien['address'])[0][0]
        dien["smalladd"] = re.findall("(\w\w[市|縣])(\w\w?\w?[區|市|鎮|鄉])",dien['address'])[0][1]
    except:
        dien["bigadd"]=0
        dien["smalladd"]=0
        
from pymongo import MongoClient,UpdateOne
client=MongoClient('192.168.1.113',27017,username='j122085',password="850605")
colle=client.rawData.canSuMe
ids=[i['name'] for i in diens]
operation=[UpdateOne({"_id":idn},{"$set":data},upsert=True) for idn,data in zip(ids,diens)]
colle.bulk_write(operation)
client.close()
```
